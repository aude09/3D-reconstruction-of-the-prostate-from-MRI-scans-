```python
import os
import pandas as pd
import SimpleITK as sitk

# ==============================
# 1️⃣ Chemins principaux
# ==============================
dicom_root = r"C:\Users\merveille\Downloads\prost"
nifti_root = r"C:\Users\merveille\Downloads\prost\nifti_out"

# Crée le dossier de sortie si nécessaire
os.makedirs(nifti_root, exist_ok=True)

# ==============================
# 2️⃣ Préparer le CSV de suivi
# ==============================
csv_records = []

# ==============================
# 3️⃣ Parcours des patients
# ==============================
for manifest in os.listdir(dicom_root):
    manifest_path = os.path.join(dicom_root, manifest)
    if not os.path.isdir(manifest_path):
        continue
    
    prostx_path = os.path.join(manifest_path, "PROSTATEx")
    if not os.path.exists(prostx_path):
        continue

    for patient_id in os.listdir(prostx_path):
        patient_path = os.path.join(prostx_path, patient_id)
        if not os.path.isdir(patient_path):
            continue

        print(f"\n📁 Patient : {patient_id}")
        
        # Crée le dossier NIfTI pour ce patient
        patient_nifti_dir = os.path.join(nifti_root, patient_id)
        os.makedirs(patient_nifti_dir, exist_ok=True)

        # ==============================
        # 4️⃣ Parcours des séries DICOM
        # ==============================
        for series_folder in os.listdir(patient_path):
            series_path = os.path.join(patient_path, series_folder)
            if not os.path.isdir(series_path):
                continue

            try:
                # Lecture DICOM et conversion NIfTI
                reader = sitk.ImageSeriesReader()
                dicom_names = reader.GetGDCMSeriesFileNames(series_path)
                if not dicom_names:
                    continue
                reader.SetFileNames(dicom_names)
                image = reader.Execute()

                # Création d'un nom simple pour le NIfTI
                series_name = series_folder.split('-')[-1]  # ou toute autre logique simple
                nifti_file = os.path.join(patient_nifti_dir, f"{series_name}.nii.gz")

                # Sauvegarde NIfTI
                sitk.WriteImage(image, nifti_file)

                # Ajout au CSV
                csv_records.append({
                    "patient_id": patient_id,
                    "series_folder": series_folder,
                    "nifti_path": nifti_file
                })

                print(f"✅ Série convertie : {series_folder} -> {nifti_file}")

            except Exception as e:
                print(f"❌ Erreur série {series_folder} : {e}")

# ==============================
# 5️⃣ Sauvegarde CSV
# ==============================
csv_file = os.path.join(nifti_root, "nifti_manifest.csv")
pd.DataFrame(csv_records).to_csv(csv_file, index=False)
print(f"\n🎯 Conversion terminée ! CSV sauvegardé : {csv_file}")
```

```python
import os
import SimpleITK as sitk
from tqdm import tqdm
base_dir = r"C:\Users\merveille\Downloads\prost\manifest-A3Y4AE4o5818678569166032044\PROSTATEx"

output_dir = r"C:\Users\merveille\Downloads\prost_nifti_T2"
os.makedirs(output_dir, exist_ok=True)

#  trouver la série T2 
def find_t2_series(patient_dir):
    for root, dirs, files in os.walk(patient_dir):
        if any(keyword in root.lower() for keyword in ["t2", "t2tse", "t2tra", "t2tsetra"]):
            dicoms = [os.path.join(root, f) for f in os.listdir(root) if f.endswith(".dcm")]
            if len(dicoms) > 5:  # au moins 5 images pour éviter les artefacts
                return root
    return None

#  Conversion 
patients = [p for p in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, p))]
print(f"🧬 {len(patients)} patients trouvés.\n")

for patient in tqdm(patients, desc="Conversion T2"):
    patient_path = os.path.join(base_dir, patient)
    t2_series_path = find_t2_series(patient_path)

    if t2_series_path is None:
        print(f" Pas de séquence T2 trouvée pour {patient}")
        continue

    try:
        # Charger les fichiers DICOM
        reader = sitk.ImageSeriesReader()
        dicom_names = reader.GetGDCMSeriesFileNames(t2_series_path)
        reader.SetFileNames(dicom_names)
        image = reader.Execute()

        # Sauvegarde au format NIfTI 
        out_path = os.path.join(output_dir, f"{patient}.nii.gz")
        sitk.WriteImage(image, out_path)
        print(f" {patient} -> {out_path}")

    except Exception as e:
        print(f" Erreur pour {patient} : {e}")

print("\n Conversion terminée ! Tous les fichiers T2 sont dans :", output_dir)
```

```python
import os

def find_pth_files(root_folder):
    """
    Cherche tous les fichiers .pth dans root_folder et ses sous-dossiers.
    """
    pth_files = []
    for dirpath, dirnames, filenames in os.walk(root_folder):
        for f in filenames:
            if f.endswith(".pth"):
                pth_files.append(os.path.join(dirpath, f))
    return pth_files

# Exemple : chercher dans tout le dossier Users
search_root = r"C:\Users\merveille"
found_files = find_pth_files(search_root)

if found_files:
    print("Fichiers .pth trouvés :")
    for f in found_files:
        print(f)
else:
    print("Aucun fichier .pth trouvé dans", search_root)
```

```python
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk
from monai.networks.nets import UNet
from ipywidgets import interact, IntSlider, VBox, Label
import torch.nn as nn

# =========================
# 1) Paramètres
# =========================
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
PATCH_D, PATCH_H, PATCH_W = 16, 128, 128

UNET_PATH = "unet3d_prostate_seg.pth"
CONV3D_PATH = "lesion_conv3d_prostatex_final.pth"

labels = ["Background", "PZ", "TZ", "CZ", "Urethra", "Other"]
cmap = plt.get_cmap("tab10", 6)


# =========================
# 2) Modèle Conv3D ProstateX
# =========================
class Conv3DClassifier(nn.Module):
    def __init__(self, in_ch=1, channels=(32, 64, 128), dropout=0.2):
        super().__init__()
        ch1, ch2, ch3 = channels
        self.encoder = nn.Sequential(
            nn.Conv3d(in_ch, ch1, 3, padding=1),
            nn.BatchNorm3d(ch1),
            nn.ReLU(inplace=True),
            nn.MaxPool3d(2),
            nn.Conv3d(ch1, ch2, 3, padding=1),
            nn.BatchNorm3d(ch2),
            nn.ReLU(inplace=True),
            nn.MaxPool3d(2),
            nn.Conv3d(ch2, ch3, 3, padding=1),
            nn.BatchNorm3d(ch3),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool3d((1, 1, 1)),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(dropout),
            nn.Linear(ch3, 1)
        )

    def forward(self, x):
        f = self.encoder(x)
        out = self.classifier(f)
        return out.squeeze(1)

# =========================
# 3) Charger les modèles
# =========================
unet_model = UNet(
    spatial_dims=3,
    in_channels=1,
    out_channels=6,
    channels=(32, 64, 128),
    strides=(2, 2),
    num_res_units=2
).to(DEVICE)
unet_model.load_state_dict(torch.load(UNET_PATH, map_location=DEVICE))
unet_model.eval()

conv3d_model = Conv3DClassifier().to(DEVICE)
conv3d_model.load_state_dict(torch.load(CONV3D_PATH, map_location=DEVICE))
conv3d_model.eval()

# =========================
# 4) Fonctions utilitaires
# =========================
def load_volume(path):
    reader = sitk.ImageSeriesReader()
    dicoms = reader.GetGDCMSeriesFileNames(path)
    reader.SetFileNames(dicoms)
    img = reader.Execute()
    arr = sitk.GetArrayFromImage(img).astype(np.float32)
    arr = (arr - arr.mean()) / (arr.std() + 1e-8)
    return arr, img


def extract_patch(arr, center, size=(PATCH_D, PATCH_H, PATCH_W)):
    D, H, W = size
    zc, yc, xc = center
    z0, y0, x0 = zc - D // 2, yc - H // 2, xc - W // 2
    z1, y1, x1 = z0 + D, y0 + H, x0 + W
    dst = np.zeros(size, dtype=arr.dtype)
    src_z0, src_y0, src_x0 = max(z0, 0), max(y0, 0), max(x0, 0)
    src_z1, src_y1, src_x1 = min(z1, arr.shape[0]), min(y1, arr.shape[1]), min(x1, arr.shape[2])
    dz0, dy0, dx0 = src_z0 - z0, src_y0 - y0, src_x0 - x0
    dz1, dy1, dx1 = dz0 + (src_z1 - src_z0), dy0 + (src_y1 - src_y0), dx0 + (src_x1 - src_x0)
    dst[dz0:dz1, dy0:dy1, dx0:dx1] = arr[src_z0:src_z1, src_y0:src_y1, src_x0:src_x1]
    return dst


def reconstruct_lesions_mask(pred_patches, centers, vol_shape):
    mask = np.zeros(vol_shape, dtype=np.uint8)
    for p, c in zip(pred_patches, centers):
        D, H, W = p.shape
        z0, y0, x0 = c[0] - D // 2, c[1] - H // 2, c[2] - W // 2
        z1, y1, x1 = z0 + D, y0 + H, x0 + W
        z0_, y0_, x0_ = max(z0, 0), max(y0, 0), max(x0, 0)
        z1_, y1_, x1_ = min(z1, vol_shape[0]), min(y1, vol_shape[1]), min(x1, vol_shape[2])
        dz0, dy0, dx0 = z0_ - z0, y0_ - y0, x0_ - x0
        dz1, dy1, dx1 = dz0 + (z1_ - z0_), dy0 + (y1_ - y0_), dx0 + (x1_ - x0_)
        mask[z0_:z1_, y0_:y1_, x0_:x1_] = np.maximum(
            mask[z0_:z1_, y0_:y1_, x0_:x1_], p[dz0:dz1, dy0:dy1, dx0:dx1]
        )
    return mask

# =========================
# 5) Pipeline complet
# =========================
def run_pipeline(volume_path):
    vol_arr, sitk_img = load_volume(volume_path)
    vol_tensor = torch.tensor(vol_arr[None, None, :, :, :], dtype=torch.float32).to(DEVICE)

    with torch.no_grad():
        zones_out = unet_model(vol_tensor)
        zones_pred = torch.argmax(zones_out, dim=1).cpu().numpy()[0]

    centers = []
    for z in range(PATCH_D // 2, vol_arr.shape[0] - PATCH_D // 2, PATCH_D):
        for y in range(PATCH_H // 2, vol_arr.shape[1] - PATCH_H // 2, PATCH_H):
            for x in range(PATCH_W // 2, vol_arr.shape[2] - PATCH_W // 2, PATCH_W):
                if np.any(zones_pred[z - PATCH_D // 2:z + PATCH_D // 2,
                                     y - PATCH_H // 2:y + PATCH_H // 2,
                                     x - PATCH_W // 2:x + PATCH_W // 2] > 0):
                    centers.append((z, y, x))

    pred_patches = []
    for c in centers:
        patch = extract_patch(vol_arr, c)
        patch_tensor = torch.tensor(patch[None, None, :, :, :], dtype=torch.float32).to(DEVICE)
        with torch.no_grad():
            pred = torch.sigmoid(conv3d_model(patch_tensor)).cpu().numpy()
        pred_patches.append((pred > 0.5).astype(np.uint8)[0])

    lesions_mask = reconstruct_lesions_mask(pred_patches, centers, vol_arr.shape)

    # === Interface interactive ===
    slice_label = Label()
    slider = IntSlider(min=0, max=vol_arr.shape[0] - 1, step=1, description="Slice")

    def plot_slice(z):
        fig, ax = plt.subplots(figsize=(6, 6))
        ax.imshow(vol_arr[z], cmap="gray", alpha=0.7)
        ax.imshow(zones_pred[z], cmap=cmap, alpha=0.4)
        ax.contour(lesions_mask[z], colors="white", linewidths=1.5)
        ax.axis("off")
        ax.set_title(f"Slice {z}")

        handles = [plt.Line2D([0], [0], color=cmap(i), lw=6) for i in range(len(labels))]
        ax.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc="upper left", title="Zones")

        plt.show()

        y, x = zones_pred[z].shape[0] // 2, zones_pred[z].shape[1] // 2
        slice_label.value = f"Centre slice {z}: zone = {labels[zones_pred[z, y, x]]}"

    interact(plot_slice, z=slider)
    display(VBox([slice_label, slider]))

    return zones_pred, lesions_mask

# =========================
# 6) Utilisation
# =========================
volume_path = r"C:\Users\merveille\Downloads\prost\manifest-A3Y4AE4o5818678569166032044\PROSTATEx\ProstateX-0341\04-17-2011-NA-MR prostaat kanker detectie NDmc MCAPRODETN-10525\13.000000-tfldynfasttra1.5x1.5t3.5sec-20400"

zones_mask, lesions_mask = run_pipeline(volume_path)
```
